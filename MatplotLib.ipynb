```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
def sinh_du_lieu():
  x = np.linspace(start=-10.0,stop=10.0, num=200)
  return x
def ham_bac_1(a, b, x):
  f = a*x + b
  return f
def ham_bac_2(a, b, c, x):

  f = a*x**2 + b*x + c
  return f
def ham_bac_3(a, b, c, d, x):
  f = a*x**3 + b*x**2 + c*x + d
  return f
def ham_cos(x):
  f = cos(x)
  return f
def ham_sin(x):
  f = sin((pi/3)*x)
  return f
```

```python
def ve_do_thi1(x,y2,y3):
  fig, ax = plt.subplots()

  ax.set_xlabel("Trục hoành - x")
  ax.set_ylabel("Trục tung - y")

  ax.plot(x, y2,label=r'$y=-2x^{2}-5x+3$')
  ax.plot(x, y3, label=r'$y=-5x + 3$')

  ax.set_title("Đồ thị phương trình bậc 2 và bậc 1")

  ax.legend()

  plt.show()  
```

```python
import matplotlib.pyplot as plt
from numpy import cos,sin,pi
def ve_nhieu_do_thi(x,y1,y2,y3):
  
  fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)

  ax1.plot(x, y1,label=r'$y=2x^{2}+3x-5$')
  ax1.plot(x, y2, label=r'$y=3x -5$')

  ax1.set_xlabel("Trục hoành - x")
  ax1.set_ylabel("Trục tung - y")
  
  ax1.legend()

  ax2.plot(x, y3, label=r'$y =cos(x)$')

  ax2.set_xlabel("Trục hoành - x")
  ax2.set_ylabel("Trục tung - y")
  
  ax2.legend()

  plt.show()
```

```python
def main():
  x = sinh_du_lieu()
  y = ham_bac_2(-2, -5,3, x)
  y2 = ham_bac_1(-5, 3, x)
  y1 = ham_bac_2(2, 3, -5, x)
  y3 = ham_cos(x)

  print("Bảng 1:")
  ve_do_thi(x,y)

  print("Bảng 2:")
  ve_do_thi1(x,y1,y2)

  print("Bảng 3: \n Đồ thị giữa phương trình bậc 2 và bậc 1 và cos")
  ve_nhieu_do_thi(x,y1,y2,y3)

if __name__ == "__main__":
  main()
```

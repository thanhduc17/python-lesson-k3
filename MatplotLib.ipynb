```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline 
```

```python

#Xây dựng hàm số
def ham_bac_2(a, b, c, x):
  f = a*x**2 + b*x + c
  return f
#sinh dữ liệu
x = np.linspace(start=-10.0,
stop=10.0, num=200)
y = ham_bac_2(2, 3, -5, x)
#Tạo đối tượng chứa hình vẽ - fig và đối tượng chứa đồ thị ax
fig, ax = plt.subplots()
#Vẽ đồ thị hàm số bậc 2 với các cặp giá trị (x, f(x) = y) tương ứng chứa trong 2 vector x và y
ax.plot(x, y)
#Hiển thị ra màn hình
plt.show()
```

```python

def ham_bac_2(a, b, c, x):
  f = a*x**2 + b*x + c
  return f
def ham_bac_1(a, b, x):
  f = a*x + b
  return f
#Tạo đối tượng chứa hình vẽ - fig và đối tượng chứa đồ thị ax
fig, ax = plt.subplots()
#sinh dữ liệu
x = np.linspace(start=-13.0,stop=10.0, num=2000)

y2 = ham_bac_2(-2, -5,3, x)
y3 = ham_bac_1(-5, 3, x)

ax.set_xlabel("Trục hoành - x")
ax.set_ylabel("Trục tung - y")
#Vẽ đồ thị hàm số bậc 2 với các cặp giá trị (x, f(x) = y) tương ứng chứa trong 2 vector x và y
ax.plot(x, y2,label=r'$y=-2x^{2}-5x+3$')
ax.plot(x, y3, label=r'$y=-5x + 3$')


ax.set_title("Đồ thị phương trình bậc 3 bậc 2 và bậc 1")

ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
from numpy import cos,sin,pi

#Tạo hình vẽ có 1 hàng 2 cột và 2đối tượng vẽ
#ax1 cho hình vẽ đồ thị hàm bậc 1và bậc 2
#ax2 cho hình vẽ hàm cosine

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)


#Chuẩn bị dữ liệu cho hàm bậc 2

x1 = np.linspace(start=-10, stop=10, num=2000)
y1 = 2*x1**2 + 3*x1 - 5

#Chuẩn bị dữ liệu cho hàm bậc 1
y2 = 3*x1 - 5

#Chuẩn bị dữ liệu cho hàm cosine

x2 = np.linspace(start=0,stop=3*pi, num=1000)
y3 = cos(x2)



#Vẽ đồ thị thứ 1

ax1.plot(x1, y1,
label=r'$y=2x^{2}+3x-5$')
ax1.plot(x1, y2, label=r'$y=3x -5$')

#Thiết lập ghi chú các trục tọa độ

ax1.set_xlabel("Trục hoành - x")
ax1.set_ylabel("Trục tung - y")

#Thông tin ghi chú đồ thị b2 và b1

ax1.legend()

#Vẽ đồ thị thứ 2

ax2.plot(x2, y3, label=r'$y =cos(x)$')

#Thiết lập ghi chú các trục tọa độ

ax2.set_xlabel("Trục hoành - x")
ax2.set_ylabel("Trục tung - y")

#Thông tin ghi chú đồ thị cos(x)

ax2.legend()
plt.show()
```

```python

#Xây dựng hàm Rosenbrock
def ham_Rosenbrock(b, x, y):
  z = (x -1)**2 + b*(y - x**2)**2
  return z

#Chuẩn bị dữ liệu

x = np.linspace(start=-2.0, stop=2.0,
num=200)
y = np.linspace(start=-1.0, stop=3.0,
num=200)
#Tạo lưới các điểm dựa trên x, y để phục vụ vẽ đồ thị 3d

x, y = np.meshgrid(x, y)
z = ham_Rosenbrock(10, x, y)

fig, ax =plt.subplots(subplot_kw={"projection":"3d"})

#Vẽ mặt Rosenbrock

rosen_surf = ax.plot_surface(x, y, z,cmap=cm.gist_heat_r,linewidth=0, antialiased=False)

#Thiết lập tham số cho trục z

ax.set_zlim(0, 200)

# Thiết lập màu dựa trên giá trị của hàm.

fig.colorbar(rosen_surf, shrink=0.5,aspect=5)

#Hiển thị hình

plt.show()
```
